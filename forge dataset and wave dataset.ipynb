```python
import mglearn
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# Forge DataSet

```python
mglearn.plots.plot_knn_classification(n_neighbors=1)
```

```python
mglearn.plots.plot_knn_classification(n_neighbors=1)
```

```python
X,y = mglearn.datasets.make_forge()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=y,random_state=30)
```

```python
print("Training features shape : ",X_train.shape)
print("Training target shape : ",y_train.shape)
print("Testing features shape : ",X_test.shape)
print("Testing target shape : ",y_test.shape)
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
```

```python
print("Accuracy of a model is : ",knn.score(X_test,y_test))
```

## Analyzing Kneighbors Classifier

```python
fig, axes = plt.subplots(1,3,figsize=(15,4))
for k,ax in zip([1,3,9],axes):
    reg = KNeighborsClassifier(n_neighbors=k).fit(X,y)
    mglearn.plots.plot_2d_separator(reg,X,fill=True,eps=0.5,ax=ax,alpha=0.4)
    mglearn.discrete_scatter(X[:,0],X[:,1],y,ax=ax)
    ax.set_title(f"{k} neigbors")
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1")
axes[0].legend(loc=3)
```

```python
mglearn.plots.plot_knn_regression(n_neighbors=1)
```

```python
mglearn.plots.plot_knn_regression(n_neighbors=3)
```

```python
X,y = mglearn.datasets.make_wave(n_samples=40)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=30)
```

```python
print("Training features shape : ",X_train.shape)
print("Training target shape : ",y_train.shape)
print("Testing features shape : ",X_test.shape)
print("Testing target shape : ",y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn = KNeighborsRegressor()
```

```python
knn.fit(X_train,y_train)
```

```python
pre = knn.predict(X_test)
pre
```

```python
print("Test score by using R^2 is : ",knn.score(X_test,y_test))
```

## Analyzing KneighborsRegressor

```python
fig, axes = plt.subplots(1,3,figsize=(15,4))
line = np.linspace(-3,3,1000).reshape(-1,1)
for k,ax in zip([1,3,9],axes):
    reg = KNeighborsRegressor(n_neighbors=k).fit(X_train,y_train)
    ax.plot(line,reg.predict(line))
    ax.plot(X_train,y_train,"^",markersize = 8)
    ax.plot(X_test,y_test,"v",markersize = 8)
    ax.set_title(f"{k} neigbors\n train score: {reg.score(X_train,y_train):.2f}\
     test score: {reg.score(X_test,y_test):.2f}")
    ax.set_xlabel("Features")
    ax.set_ylabel("Target")
axes[1].legend(["Model Prediction","Training data/target",
               "Test data/target"],loc="best")
```
