```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = load_breast_cancer()
```

```python
X = df.data
y = df.target
print("Input shape is : ",X.shape)
print("Target shape is : ",y.shape)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,stratify=y,random_state=30)
```

```python
print("Training features shape : ",X_train.shape)
print("Training target shape : ",y_train.shape)
print("Testing features shape : ",X_test.shape)
print("Testing target shape : ",y_test.shape)
```

## visualizing best value of k

```python
training_accuracy = []
testing_accuracy = []
neihbors_setting = range(1,11)
```

```python
for k in neihbors_setting:
    knn = KNeighborsClassifier(n_neighbors=k).fit(X_train,y_train)
    p1 = knn.score(X_train,y_train)
    p2 = knn.score(X_test,y_test)
    training_accuracy.append(p1)
    testing_accuracy.append(p2)
```

```python
plt.figure(figsize=(5,5))
plt.plot(neihbors_setting,training_accuracy,label="training accuracy")
plt.plot(neihbors_setting,testing_accuracy,label="testing accuracy")
plt.xlabel("K_values")
plt.ylabel("Performance Measure")
plt.legend()
plt.show()
```

# Best value for k is 6

```python
knn = KNeighborsClassifier(n_neighbors=6)
knn.fit(X_train,y_train)
acc = knn.score(X_test,y_test)
print("Accuracy of a model is : ",acc)
```
