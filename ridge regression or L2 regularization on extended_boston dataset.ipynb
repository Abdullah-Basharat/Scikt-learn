1) Ridge Regression:
* α is a non-negative value.
* It controls the amount of regularization applied to the model.
* As α increases, the regularization strength increases.
* When α is 0, Ridge regression becomes equivalent to ordinary least squares regression (no regularization).
* As α tends towards infinity, the coefficients tend towards zero.
2) Lasso Regression:
* α is a non-negative value.
* It also controls the amount of regularization applied to the model.
* Similar to Ridge, as α increases, the regularization strength increases.
* One key difference is that Lasso regression can lead to sparse models (i.e., some coefficients are exactly zero), effectively performing feature selection.
* Lasso's penalty term has a tendency to force some coefficients to zero, effectively eliminating them from the model.
* Lasso is particularly useful when dealing with high-dimensional datasets where feature selection is important

```python
import mglearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge
```

```python
X,y = mglearn.datasets.load_extended_boston()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)
```

# Model 1 (Linear Regression)

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
print("Test score on training data: ",lm.score(X_train,y_train))
print("Test score on testing data: ",lm.score(X_test,y_test))
```

# Model 2 (Ridge Regression with alpha = 1.0)

```python
ridge1 = Ridge()
```

```python
ridge1.fit(X_train,y_train)
```

```python
print("Test score on training data: ",ridge1.score(X_train,y_train))
print("Test score on testing data: ",ridge1.score(X_test,y_test))
```

# Model 3 (Ridge Regression with alpha = 10)

```python
ridge10 = Ridge(alpha=10)
```

```python
ridge10.fit(X_train,y_train)
```

```python
print("Test score on training data: ",ridge10.score(X_train,y_train))
print("Test score on testing data: ",ridge10.score(X_test,y_test))
```

# Model 4 (Ridge Regression with alpha = 0.1)

```python
ridge01 = Ridge(alpha=0.1)
```

```python
ridge01.fit(X_train,y_train)
```

```python
print("Test score on training data: ",ridge01.score(X_train,y_train))
print("Test score on testing data: ",ridge01.score(X_test,y_test))
```

## Beahvior of alpha
* When alpha is large, it helps to select coefficients near to zero so solpe is minimum
* When alpha is very small, it makes the ridge model similar to linear regression model.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(lm.coef_,"s",label="Linear Regression")
plt.plot(ridge1.coef_,"v",label="Ridge1 Regression")
plt.plot(ridge10.coef_,"^",label="Ridge10 Regression")
plt.plot(ridge01.coef_,"o",label="Ridge01 Regression")
plt.xlabel("Coefficient index")
plt.ylabel("Coefficient Magnitude")
plt.ylim(-25,25)
plt.legend(loc="best")
plt.show()
```
